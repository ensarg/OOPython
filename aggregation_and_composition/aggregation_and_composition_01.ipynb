```python
# Aggregation

class Salary:
    def __init__(self, pay):
        self.pay = pay

    def get_total(self):
        return self.pay * 12

class Employee:
    def __init__(self, pay, bonus):
        self.pay = pay
        self.bonus = bonus
        self.annual_salary = Salary(self.pay) # Aggregation: Employee has a Salary

    def display_total_compensation(self):
        total_compensation = self.annual_salary.get_total() + self.bonus
        print(f"Total Employee Compensation (Aggregation): ${total_compensation}")

# Composition

class Engine:
    def __init__(self, horsepower):
        self.horsepower = horsepower

    def start(self):
        print("Engine starting...")

class Car:
    def __init__(self, make, model, horsepower):
        self.make = make
        self.model = model
        self.engine = Engine(horsepower) # Composition: Car has an Engine

    def start_car(self):
        print(f"Starting the {self.make} {self.model}")
        self.engine.start()

# Demonstrate Aggregation
employee1 = Employee(50000, 5000)
employee1.display_total_compensation()

```

```python
# Composition

class Engine:
    def __init__(self, horsepower):
        self.horsepower = horsepower

    def start(self):
        print("Engine starting...")

class Car:
    def __init__(self, make, model, horsepower):
        self.make = make
        self.model = model
        self.engine = Engine(horsepower) # Composition: Car has an Engine

    def start_car(self):
        print(f"Starting the {self.make} {self.model}")
        self.engine.start()



# Demonstrate Composition
car1 = Car("Toyota", "Camry", 180)
car1.start_car()
```

In the code above:

**Aggregation** is demonstrated with the `Employee` and `Salary` classes. An `Employee` "has a" `Salary`. The `Salary` object can exist independently of the `Employee` object. If an `Employee` object is destroyed, the `Salary` object can still exist.

**Composition** is demonstrated with the `Car` and `Engine` classes. A `Car` "has an" `Engine`. The `Engine` object's lifecycle is dependent on the `Car` object. If a `Car` object is destroyed, the `Engine` object that is part of it is also typically destroyed (or at least no longer accessible through the `Car` object).
