```python
def divide_numbers(a, b):
    try:
        # Try block: code that may raise an exception
        result = a / b
    except ZeroDivisionError:
        # Handles division by zero error
        print("Error: Cannot divide by zero.")
    except TypeError:
        # Handles invalid type error (e.g., dividing string by number)
        print("Error: Both values must be numbers.")
    else:
        # Executes if no exception occurs
        print(f"Division successful! Result = {result}")
    finally:
        # Always executes, regardless of exception
        print("Execution completed.\n")


# Test cases
divide_numbers(10, 2)     # No error
divide_numbers(10, 0)     # ZeroDivisionError
divide_numbers(10, "a")   # TypeError
```

try block: Contains code that might throw an exception.

except block(s): Used to handle specific exceptions.

else block: Runs only if no exception occurs.

finally block: Runs no matter what â€” used for cleanup (like closing files or connections).
