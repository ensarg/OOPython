```python
#inheritance
# Base class
class Animal:
    def __init__(self, name):
        self.name = name

    # A general method to make a sound
    def make_sound(self):
        print(f"{self.name} makes a sound.")

    # Example of method overloading using default arguments
    def eat(self, food=None):
        if food is None:
            print(f"{self.name} is eating.")
        else:
            print(f"{self.name} is eating {food}.")

# Derived class
class Cat(Animal):
    def __init__(self, name, breed):
        # Inheriting properties from the base class
        super().__init__(name)
        self.breed = breed

    # Overriding the make_sound() method
    def make_sound(self):
        print(f"{self.name} says: Meow!")

    # Overloading-like behavior (Python doesn’t support true overloading)
    def eat(self, food=None, amount=None):
        if food is None and amount is None:
            print(f"{self.name} is nibbling on some food.")
        elif amount is None:
            print(f"{self.name} is eating {food}.")
        else:
            print(f"{self.name} is eating {amount} of {food}.")

# --- Example usage ---
if __name__ == "__main__":
    # Creating an Animal object
    animal = Animal("Generic Animal")
    animal.make_sound()
    animal.eat()
    animal.eat("grass")

    print("\n--- Cat Class Example ---")
    # Creating a Cat object
    cat = Cat("Whiskers", "Persian")
    cat.make_sound()     # Overridden method
    cat.eat()            # Overloaded method variant 1
    cat.eat("fish")      # Overloaded method variant 2
    cat.eat("fish", "a lot")  # Overloaded method variant 3
```

Inheritance:
The Cat class inherits from the Animal class using class Cat(Animal).

Method Overriding:
The Cat class overrides the make_sound() method to provide its own version.

Method Overloading (Simulated):
Python doesn’t support true overloading like Java or C++, but you can simulate it using default parameters or *args/**kwargs.
The eat() method in both classes demonstrates this.
