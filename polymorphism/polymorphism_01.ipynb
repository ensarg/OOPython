```python
class Animal:
    def speak(self):
        return "Some generic animal sound"

class Dog(Animal):
    def speak(self):
        return "Woof! Woof!"

class Cat(Animal):
    def speak(self):
        return "Meow!"

class Cow(Animal):
    def speak(self):
        return "Moo!"

# Polymorphism in action
animals = [Dog(), Cat(), Cow()]

for animal in animals:
    print(f"{animal.__class__.__name__} says: {animal.speak()}")
```

All subclasses (Dog, Cat, Cow) override the speak() method of the base class Animal.

When you call animal.speak(), Python automatically uses the appropriate method depending on the object type — even though you’re calling the same method name.
