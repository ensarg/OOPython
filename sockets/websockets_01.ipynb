### WebSocket Server Example

```python
import asyncio
import websockets

async def echo(websocket):
    async for message in websocket:
        print(f"Received message from client: {message}")
        await websocket.send(f"Echo: {message}")

async def main():
    # Start the WebSocket server on localhost, port 8765
    async with websockets.serve(echo, "localhost", 8765):
        print("WebSocket server started on ws://localhost:8765")
        await asyncio.Future()  # Run forever

# To run the server, uncomment the line below and execute this cell:
#asyncio.run(main())
await main()
```

### WebSocket Client Example

```python
import asyncio
import websockets

async def send_message():
    # Connect to the WebSocket server running on localhost, port 8765
    async with websockets.connect("ws://localhost:8765") as websocket:
        message = "Hello, WebSocket Server!"
        print(f"Sending message to server: {message}")
        await websocket.send(message)

        response = await websocket.recv()
        print(f"Received response from server: {response}")

# To run the client, uncomment the line below and execute this cell:
#asyncio.run(send_message())
await send_message()
```

**Instructions:**

1.  Run the server code cell first (uncomment the `asyncio.run(main())` line). This will start the server and it will keep running.
2.  In a separate execution, run the client code cell (uncomment the `asyncio.run(send_message())` line). The client will connect to the server, send a message, and print the response.

Remember that in a real-world scenario, you would typically run the server and client in separate processes or machines.
