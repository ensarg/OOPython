```python
class Student:
    # Static (class) variable
    school_name = "Global High School"

    def __init__(self, name, age):
        # Instance variables
        self.name = name
        self.age = age

    def show(self):
        print(f"Student Name: {self.name}, Age: {self.age}, School: {Student.school_name}")

    @staticmethod
    def school_info():
        print("This is a static method.")
        print(f"All students study at {Student.school_name}")


# Creating objects
s1 = Student("Alice", 15)
s2 = Student("Bob", 16)

# Accessing instance method
s1.show()
s2.show()

# Accessing static variable through class and instance
print("\nAccessing static variable directly through class:")
print(Student.school_name)

print("\nAccessing static variable through object:")
print(s1.school_name)

# Calling static method
print("\nCalling static method:")
Student.school_info()
```

Static variable (school_name) belongs to the class, not to any particular object.

Changing it via the class affects all instances.

Instance variables (self.name, self.age) belong to individual objects.

Static method (school_info) does not depend on instance or class variables directly — it’s used for functionality that is logically related to the class.
