```python
class MyClass:
    # Static variable (class variable)
    static_variable = "I am a static variable"

    def __init__(self, instance_variable):
        # Instance variable
        self.instance_variable = instance_variable

    # Static method
    @staticmethod
    def static_method(message):
        print(f"Inside static method: {message}")
        print(f"Accessing static variable from static method: {MyClass.static_variable}")

    # Instance method
    def instance_method(self):
        print(f"Inside instance method: {self.instance_variable}")
        print(f"Accessing static variable from instance method: {MyClass.static_variable}")

# Accessing the static variable directly using the class name
print(f"Accessing static variable directly: {MyClass.static_variable}")

# Calling the static method directly using the class name
MyClass.static_method("Hello from static method!")

# Creating instances of the class
obj1 = MyClass("Instance 1 data")
obj2 = MyClass("Instance 2 data")

# Accessing static variable through instances (not recommended, but possible)
print(f"Accessing static variable through obj1: {obj1.static_variable}")
print(f"Accessing static variable through obj2: {obj2.static_variable}")

# Calling instance methods
obj1.instance_method()
obj2.instance_method()

# Demonstrating that changing the static variable through one instance affects all instances
obj1.static_variable = "Static variable changed via obj1"
print(f"Accessing static variable directly after change via obj1: {MyClass.static_variable}")
print(f"Accessing static variable through obj2 after change via obj1: {obj2.static_variable}")

# Note: Changing the static variable via the class name is the recommended way
MyClass.static_variable = "Static variable changed via class name"
print(f"Accessing static variable directly after change via class name: {MyClass.static_variable}")
print(f"Accessing static variable through obj1 after change via class name: {obj1.static_variable}")
print(f"Accessing static variable through obj2 after change via class name: {obj2.static_variable}")
```
